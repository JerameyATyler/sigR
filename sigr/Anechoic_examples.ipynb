# Anechoic

The `Anechoic` class extends `pytorch.utils.data.Dataset` and simplifies working with datasets in `pytorch`.

## `Anechoic`

The `__len__` and `__getitem__` functions are overriden to allow `Anechoic` to behave like a `list`, see below.

```python
from Anechoic import Anechoic

ani = Anechoic('./data', ttv='test')
len(ani)
```
